```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
Y = dataset.iloc[:,-1].values
```

# Linear Regression Prediction

```python
from sklearn.linear_model import LinearRegression
lin_reg_1 = LinearRegression()
lin_reg_1.fit(X,Y)
```

```python
lin_reg_1.predict([[6.5]])
```

```python
plt.scatter(X,Y,color = "red")
plt.plot(X, lin_reg_1.predict(X),color = "blue")
plt.title("linear regression prediction")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

# Polynomial Regression Prediction

```python
from sklearn.preprocessing import PolynomialFeatures
X_poly = PolynomialFeatures(degree = 3)
X_features = X_poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_features,Y)
```

```python
poly_reg.predict(X_poly.fit_transform([[6.5]]))
```

```python
plt.scatter(X,Y, color = "red")
plt.plot(X, poly_reg.predict(X_features), color = "Blue")
plt.title("Polynomial regression results")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
